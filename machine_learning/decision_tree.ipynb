```python
import pandas as pd
from PIL import Image
```

```python
iris = pd.read_csv('D:\studydata\machine_bild_learn\AI\machine_learning\Iris数据集\iris.csv')
iris.head()
```

```python
iris.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(iris.dropna(), hue='Species', size=2.5)
```

```python
plt.figure(figsize=(10, 10))
for column_index, column in enumerate(iris.columns):
    if column == 'Species':
        continue
    plt.subplot(2, 2, column_index + 1)
    sns.violinplot(x='Species', y=column, data=iris)
```

```python
from sklearn.model_selection import train_test_split
all_inputs = iris[['Sepal.Length', 'Sepal.Width',
                     'Petal.Length', 'Petal.Width']].values
all_classes = iris['Species'].values
(train_inputs, test_inputs, train_classes, test_classes) = train_test_split(all_inputs, all_classes, train_size=0.7, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
# Create the classifier
decision_tree_classifier = DecisionTreeClassifier()
# Train the classifier on the training set
decision_tree_classifier.fit(train_inputs, train_classes)
# Validate the classifier on the testing set using classification accuracy
decision_tree_classifier.score(test_inputs, test_classes)

```

```python
# Predict the classes of new, unseen data

decision_tree_classifier.predict(test_inputs[0:1])
```
